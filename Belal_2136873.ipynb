# • Importing All Libraries We Need:

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Load The Images:

```python
apple = Image.open("Apple.jpg")
orange = Image.open("Orange.jpg")
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Convert Images To Numpy Arrays:

```python
apple_array = np.array(apple)
orange_array = np.array(orange)
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Create a Alpha Mask (left: 1, right: 0) ==> Alpha-Left:

```python
width, height = apple.size
mask = np.linspace(1, 0, width)  # 1 on the left side of the image and slowly fades to 0 on the right.
mask = np.tile(mask, (height, 1))  # Make it the size of the image. (from 1D array to 2D array.).
mask = np.expand_dims(mask, axis=-1)  # adds an extra dimension alpha. (from a 2D array (height, width) to a 3D array (height, width, 1)).
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Assign The 3D Array To Apple Mask And Orange Mask:

```python
apple_mask = mask # 1 to 0.
orange_mask = 1 - mask  #0 to 1.
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Perform Blending:

```python
blended_array = (apple_array * apple_mask + orange_array * orange_mask).astype(np.uint8)
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Convert The Numpy Array Image:

```python
blended_image = Image.fromarray(blended_array)
```

# ــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــــ

# • Save And Show The Blended Image:

```python
blended_image.save("blended_image.jpg")

plt.imshow(blended_image, cmap='gray')
```
